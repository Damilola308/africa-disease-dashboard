An interactive Plotly Dash web application visualising disease burden
across African Union member states. Data is simulated to reflect
realistic epidemiological patterns based on publicly available
Africa CDC and WHO disease surveillance reports.

```python
import dash
from dash import dcc, html, Input, Output, dash_table
import plotly.express as px
import plotly.graph_objects as go
import pandas as pd
import numpy as np
```

```python



# ── 1. SIMULATE REALISTIC DISEASE BURDEN DATA ─────────────────────────────────

np.random.seed(42)

countries = [
    'Nigeria', 'Ethiopia', 'Egypt', 'South Africa', 'Kenya',
    'Tanzania', 'Ghana', 'Uganda', 'Algeria', 'Sudan',
    'Mozambique', 'Madagascar', 'Cameroon', 'Ivory Coast', 'Niger',
    'Burkina Faso', 'Mali', 'Malawi', 'Zambia', 'Senegal',
    'Zimbabwe', 'Rwanda', 'Guinea', 'Benin', 'Tunisia',
    'Morocco', 'Sierra Leone', 'Togo', 'Libya', 'Congo'
]

country_codes = [
    'NGA', 'ETH', 'EGY', 'ZAF', 'KEN',
    'TZA', 'GHA', 'UGA', 'DZA', 'SDN',
    'MOZ', 'MDG', 'CMR', 'CIV', 'NER',
    'BFA', 'MLI', 'MWI', 'ZMB', 'SEN',
    'ZWE', 'RWA', 'GIN', 'BEN', 'TUN',
    'MAR', 'SLE', 'TGO', 'LBY', 'COG'
]

regions = [
    'West Africa', 'East Africa', 'North Africa', 'Southern Africa', 'East Africa',
    'East Africa', 'West Africa', 'East Africa', 'North Africa', 'East Africa',
    'Southern Africa', 'East Africa', 'Central Africa', 'West Africa', 'West Africa',
    'West Africa', 'West Africa', 'Southern Africa', 'Southern Africa', 'West Africa',
    'Southern Africa', 'East Africa', 'West Africa', 'West Africa', 'North Africa',
    'North Africa', 'West Africa', 'West Africa', 'North Africa', 'Central Africa'
]

populations_M = [
    220, 125, 105, 60, 55,
    63, 33, 48, 45, 45,
    33, 28, 27, 27, 25,
    22, 22, 20, 19, 17,
    16, 14, 13, 13, 12,
    37, 8, 8, 7, 6
]

years = list(range(2018, 2024))
diseases = ['Malaria', 'Cholera', 'Mpox', 'Tuberculosis', 'Meningitis']

# Base case rates per 100,000 population (approximate real-world values)
base_rates = {
    'Malaria': {
        'Nigeria': 18000, 'Ethiopia': 8000, 'Egypt': 50, 'South Africa': 200,
        'Kenya': 6000, 'Tanzania': 9000, 'Ghana': 7500, 'Uganda': 11000,
        'Algeria': 30, 'Sudan': 5000, 'Mozambique': 12000, 'Madagascar': 8500,
        'Cameroon': 9500, 'Ivory Coast': 8800, 'Niger': 14000,
        'Burkina Faso': 15000, 'Mali': 13000, 'Malawi': 11500, 'Zambia': 10000,
        'Senegal': 7000, 'Zimbabwe': 5000, 'Rwanda': 4000, 'Guinea': 12000,
        'Benin': 9000, 'Tunisia': 20, 'Morocco': 15, 'Sierra Leone': 13000,
        'Togo': 8500, 'Libya': 25, 'Congo': 10000
    },
    'Cholera': {
        'Nigeria': 120, 'Ethiopia': 200, 'Egypt': 10, 'South Africa': 80,
        'Kenya': 90, 'Tanzania': 150, 'Ghana': 60, 'Uganda': 110,
        'Algeria': 5, 'Sudan': 300, 'Mozambique': 250, 'Madagascar': 180,
        'Cameroon': 130, 'Ivory Coast': 80, 'Niger': 200,
        'Burkina Faso': 170, 'Mali': 190, 'Malawi': 220, 'Zambia': 160,
        'Senegal': 70, 'Zimbabwe': 400, 'Rwanda': 40, 'Guinea': 210,
        'Benin': 90, 'Tunisia': 3, 'Morocco': 2, 'Sierra Leone': 230,
        'Togo': 100, 'Libya': 8, 'Congo': 280
    },
    'Mpox': {
        'Nigeria': 15, 'Ethiopia': 2, 'Egypt': 1, 'South Africa': 8,
        'Kenya': 3, 'Tanzania': 2, 'Ghana': 4, 'Uganda': 5,
        'Algeria': 1, 'Sudan': 3, 'Mozambique': 2, 'Madagascar': 1,
        'Cameroon': 12, 'Ivory Coast': 6, 'Niger': 4,
        'Burkina Faso': 5, 'Mali': 4, 'Malawi': 2, 'Zambia': 3,
        'Senegal': 3, 'Zimbabwe': 2, 'Rwanda': 2, 'Guinea': 8,
        'Benin': 5, 'Tunisia': 1, 'Morocco': 1, 'Sierra Leone': 6,
        'Togo': 4, 'Libya': 1, 'Congo': 25
    },
    'Tuberculosis': {
        'Nigeria': 219, 'Ethiopia': 140, 'Egypt': 12, 'South Africa': 615,
        'Kenya': 292, 'Tanzania': 253, 'Ghana': 148, 'Uganda': 201,
        'Algeria': 70, 'Sudan': 180, 'Mozambique': 558, 'Madagascar': 233,
        'Cameroon': 175, 'Ivory Coast': 165, 'Niger': 98,
        'Burkina Faso': 122, 'Mali': 130, 'Malawi': 181, 'Zambia': 350,
        'Senegal': 119, 'Zimbabwe': 292, 'Rwanda': 58, 'Guinea': 176,
        'Benin': 92, 'Tunisia': 30, 'Morocco': 100, 'Sierra Leone': 298,
        'Togo': 110, 'Libya': 42, 'Congo': 350
    },
    'Meningitis': {
        'Nigeria': 45, 'Ethiopia': 20, 'Egypt': 5, 'South Africa': 8,
        'Kenya': 15, 'Tanzania': 18, 'Ghana': 30, 'Uganda': 22,
        'Algeria': 4, 'Sudan': 35, 'Mozambique': 12, 'Madagascar': 10,
        'Cameroon': 28, 'Ivory Coast': 25, 'Niger': 60,
        'Burkina Faso': 55, 'Mali': 50, 'Malawi': 15, 'Zambia': 18,
        'Senegal': 28, 'Zimbabwe': 10, 'Rwanda': 8, 'Guinea': 35,
        'Benin': 30, 'Tunisia': 3, 'Morocco': 3, 'Sierra Leone': 32,
        'Togo': 28, 'Libya': 4, 'Congo': 40
    }
}

# Build long-format dataframe
records = []
for country, code, region, pop in zip(countries, country_codes, regions, populations_M):
    for disease in diseases:
        for year in years:
            base = base_rates[disease][country]
            # Add year-over-year trend + noise
            trend = np.random.uniform(-0.08, 0.05)
            year_factor = (1 + trend) ** (year - 2018)
            noise = np.random.uniform(0.85, 1.15)
            rate = max(0, round(base * year_factor * noise, 1))
            cases = int(rate * pop / 100)
            records.append({
                'Country': country,
                'ISO': code,
                'Region': region,
                'Population_M': pop,
                'Disease': disease,
                'Year': year,
                'Rate_per_100k': rate,
                'Estimated_Cases': cases
            })

df = pd.DataFrame(records)

```

```python
# ── 2. APP LAYOUT ──────────────────────────────────────────────────────────────

app = dash.Dash(__name__, title="Africa CDC Disease Surveillance Dashboard")

NAVY   = "#1B3A6B"
GREEN  = "#008000"
TURQUOISE  = "#40E0D0"
YELLOW  = "#FFFF00"
ACCENT = "#2563EB"
BG     = "#F8FAFC"
CARD   = "#FFFFFF"

app.layout = html.Div(style={'backgroundColor': BG, 'fontFamily': 'Arial, sans-serif', 'minHeight': '100vh'}, children=[

    # Header
    html.Div(style={
        'backgroundColor': NAVY, 'padding': '24px 40px',
        'marginBottom': '24px'
    }, children=[
        html.H1("🌍 Africa CDC Disease Surveillance Dashboard",
                style={'color': 'white', 'margin': 0, 'fontSize': '26px', 'fontWeight': 'bold'}),
        html.P("Interactive visualisation of disease burden across African Union member states (2018–2023)",
               style={'color': '#93C5FD', 'margin': '6px 0 0 0', 'fontSize': '14px'}),
        html.P("Author: Opeyemi Ogunbona   |   Data: Simulated based on Africa CDC & WHO surveillance reports",
               style={'color': '#FFFF00', 'margin': '4px 0 0 0', 'fontSize': '12px'}),
    ]),

    # Filters Row
    html.Div(style={'display': 'flex', 'gap': '20px', 'padding': '0 40px', 'marginBottom': '24px', 'flexWrap': 'wrap'}, children=[

        html.Div(style={'flex': '1', 'minWidth': '200px'}, children=[
            html.Label("Disease", style={'fontWeight': 'bold', 'color': NAVY, 'fontSize': '13px'}),
            dcc.Dropdown(
                id='disease-filter',
                options=[{'label': d, 'value': d} for d in diseases],
                value='Malaria',
                clearable=False,
                style={'marginTop': '6px'}
            )
        ]),

        html.Div(style={'flex': '1', 'minWidth': '200px'}, children=[
            html.Label("Year", style={'fontWeight': 'bold', 'color': NAVY, 'fontSize': '13px'}),
            dcc.Dropdown(
                id='year-filter',
                options=[{'label': str(y), 'value': y} for y in years],
                value=2023,
                clearable=False,
                style={'marginTop': '6px'}
            )
        ]),

        html.Div(style={'flex': '2', 'minWidth': '280px'}, children=[
            html.Label("Countries (for trend chart)", style={'fontWeight': 'bold', 'color': NAVY, 'fontSize': '13px'}),
            dcc.Dropdown(
                id='country-filter',
                options=[{'label': c, 'value': c} for c in sorted(countries)],
                value=['Nigeria', 'South Africa', 'Kenya', 'Ethiopia', 'Ghana'],
                multi=True,
                style={'marginTop': '6px'}
            )
        ]),
    ]),

    # Summary Cards
    html.Div(id='summary-cards', style={
        'display': 'flex', 'gap': '16px', 'padding': '0 40px',
        'marginBottom': '24px', 'flexWrap': 'wrap'
    }),

    # Charts Row 1 — Map + Bar
    html.Div(style={'display': 'flex', 'gap': '20px', 'padding': '0 40px', 'marginBottom': '20px', 'flexWrap': 'wrap'}, children=[

        html.Div(style={'flex': '2', 'minWidth': '340px', 'backgroundColor': CARD,
                        'borderRadius': '8px', 'padding': '20px', 'boxShadow': '0 1px 4px rgba(0,0,0,0.08)'}, children=[
            html.H3("Disease Rate by Country (per 100,000)", style={'color': TURQUOISE, 'fontSize': '15px', 'marginTop': 0}),
            dcc.Graph(id='choropleth-map', style={'height': '380px'})
        ]),

        html.Div(style={'flex': '1', 'minWidth': '280px', 'backgroundColor': CARD,
                        'borderRadius': '8px', 'padding': '20px', 'boxShadow': '0 1px 4px rgba(0,0,0,0.08)'}, children=[
            html.H3("Top 10 Countries by Case Rate", style={'color': NAVY, 'fontSize': '15px', 'marginTop': 0}),
            dcc.Graph(id='bar-chart', style={'height': '380px'})
        ]),
    ]),

    # Charts Row 2 — Trend + Region
    html.Div(style={'display': 'flex', 'gap': '20px', 'padding': '0 40px', 'marginBottom': '20px', 'flexWrap': 'wrap'}, children=[

        html.Div(style={'flex': '2', 'minWidth': '340px', 'backgroundColor': CARD,
                        'borderRadius': '8px', 'padding': '20px', 'boxShadow': '0 1px 4px rgba(0,0,0,0.08)'}, children=[
            html.H3("Disease Trend Over Time (Selected Countries)", style={'color': NAVY, 'fontSize': '15px', 'marginTop': 0}),
            dcc.Graph(id='trend-chart', style={'height': '340px'})
        ]),

        html.Div(style={'flex': '1', 'minWidth': '280px', 'backgroundColor': CARD,
                        'borderRadius': '8px', 'padding': '20px', 'boxShadow': '0 1px 4px rgba(0,0,0,0.08)'}, children=[
            html.H3("Average Rate by Region", style={'color': NAVY, 'fontSize': '15px', 'marginTop': 0}),
            dcc.Graph(id='region-chart', style={'height': '340px'})
        ]),
    ]),

    # Data Table
    html.Div(style={'padding': '0 40px', 'marginBottom': '40px'}, children=[
        html.Div(style={'backgroundColor': CARD, 'borderRadius': '8px', 'padding': '20px',
                        'boxShadow': '0 1px 4px rgba(0,0,0,0.08)'}, children=[
            html.H3("Country Data Table", style={'color': NAVY, 'fontSize': '15px', 'marginTop': 0}),
            html.Div(id='data-table')
        ])
    ]),

    # Footer
    html.Div(style={'backgroundColor': NAVY, 'padding': '16px 40px', 'textAlign': 'center'}, children=[
        html.P("Data Note: Disease burden figures are simulated using epidemiological baselines from Africa CDC and WHO surveillance reports. For educational purposes only.",
               style={'color': '#FFFF00', 'fontSize': '11px', 'margin': 0})
    ])
])

```

```python
# ── 3. CALLBACKS ───────────────────────────────────────────────────────────────

@app.callback(
    Output('summary-cards', 'children'),
    Output('choropleth-map', 'figure'),
    Output('bar-chart', 'figure'),
    Output('trend-chart', 'figure'),
    Output('region-chart', 'figure'),
    Output('data-table', 'children'),
    Input('disease-filter', 'value'),
    Input('year-filter', 'value'),
    Input('country-filter', 'value'),
)
def update_dashboard(disease, year, selected_countries):
    # Filter data
    filtered = df[(df['Disease'] == disease) & (df['Year'] == year)]
    trend_data = df[(df['Disease'] == disease) & (df['Country'].isin(selected_countries or ['Nigeria']))]

    # ── Summary Cards
    total_cases = filtered['Estimated_Cases'].sum()
    avg_rate    = filtered['Rate_per_100k'].mean()
    highest     = filtered.nlargest(1, 'Rate_per_100k').iloc[0]
    lowest      = filtered[filtered['Rate_per_100k'] > 0].nsmallest(1, 'Rate_per_100k').iloc[0]

    def card(title, value, sub, color):
        return html.Div(style={
            'backgroundColor': CARD, 'borderRadius': '8px', 'padding': '16px 20px',
            'flex': '1', 'minWidth': '160px', 'boxShadow': '0 1px 4px rgba(0,0,0,0.08)',
            'borderTop': f'4px solid {color}'
        }, children=[
            html.P(title, style={'color': '#6B7280', 'fontSize': '12px', 'margin': '0 0 6px 0', 'fontWeight': 'bold'}),
            html.P(value, style={'color': NAVY, 'fontSize': '22px', 'fontWeight': 'bold', 'margin': '0 0 4px 0'}),
            html.P(sub,   style={'color': '#9CA3AF', 'fontSize': '11px', 'margin': 0})
        ])

    cards = [
        card("Total Estimated Cases", f"{total_cases:,.0f}", f"{disease} across Africa — {year}", ACCENT),
        card("Avg Rate per 100k", f"{avg_rate:,.1f}", "Unweighted country average", "#10B981"),
        card("Highest Burden", highest['Country'], f"{highest['Rate_per_100k']:,.1f} per 100k", "#EF4444"),
        card("Lowest Burden", lowest['Country'],  f"{lowest['Rate_per_100k']:,.1f} per 100k", "#F59E0B"),
    ]

    # ── Choropleth Map
    map_fig = px.choropleth(
        filtered, locations='ISO', color='Rate_per_100k',
        hover_name='Country',
        hover_data={'ISO': False, 'Rate_per_100k': ':.1f', 'Estimated_Cases': ':,.0f'},
        color_continuous_scale='YlOrRd',
        scope='africa',
        labels={'Rate_per_100k': 'Rate per 100k'},
        title=f'{disease} Rate per 100,000 Population ({year})'
    )
    map_fig.update_layout(
        margin=dict(l=0, r=0, t=40, b=0),
        coloraxis_colorbar=dict(title='Rate/100k', thickness=12),
        paper_bgcolor='white', geo_bgcolor='#EFF6FF'
    )

    # ── Bar Chart — Top 10
    top10 = filtered.nlargest(10, 'Rate_per_100k').sort_values('Rate_per_100k')
    bar_fig = go.Figure(go.Bar(
        x=top10['Rate_per_100k'], y=top10['Country'],
        orientation='h',
        marker_color=ACCENT,
        text=top10['Rate_per_100k'].apply(lambda x: f'{x:,.1f}'),
        textposition='outside'
    ))
    bar_fig.update_layout(
        margin=dict(l=0, r=60, t=10, b=0),
        xaxis_title='Rate per 100,000',
        paper_bgcolor='white', plot_bgcolor='white',
        xaxis=dict(showgrid=True, gridcolor='#F1F5F9')
    )

    # ── Trend Chart
    trend_fig = go.Figure()
    colors = px.colors.qualitative.Bold
    for i, country in enumerate(selected_countries or ['Nigeria']):
        cdata = trend_data[trend_data['Country'] == country]
        trend_fig.add_trace(go.Scatter(
            x=cdata['Year'], y=cdata['Rate_per_100k'],
            mode='lines+markers', name=country,
            line=dict(color=colors[i % len(colors)], width=2.5),
            marker=dict(size=6)
        ))
    trend_fig.update_layout(
        margin=dict(l=0, r=0, t=10, b=0),
        xaxis_title='Year', yaxis_title='Rate per 100,000',
        paper_bgcolor='white', plot_bgcolor='white',
        xaxis=dict(showgrid=True, gridcolor='#F1F5F9', dtick=1),
        yaxis=dict(showgrid=True, gridcolor='#F1F5F9'),
        legend=dict(orientation='h', yanchor='bottom', y=1.02, xanchor='right', x=1)
    )

    # ── Region Chart
    region_avg = filtered.groupby('Region')['Rate_per_100k'].mean().sort_values(ascending=True).reset_index()
    region_fig = go.Figure(go.Bar(
        x=region_avg['Rate_per_100k'], y=region_avg['Region'],
        orientation='h',
        marker_color='#10B981',
        text=region_avg['Rate_per_100k'].apply(lambda x: f'{x:,.1f}'),
        textposition='outside'
    ))
    region_fig.update_layout(
        margin=dict(l=0, r=60, t=10, b=0),
        xaxis_title='Avg Rate per 100,000',
        paper_bgcolor='white', plot_bgcolor='white',
        xaxis=dict(showgrid=True, gridcolor='#F1F5F9')
    )

    # ── Data Table
    table_df = filtered[['Country', 'Region', 'Population_M', 'Rate_per_100k', 'Estimated_Cases']]\
        .sort_values('Rate_per_100k', ascending=False)\
        .rename(columns={
            'Population_M': 'Population (M)',
            'Rate_per_100k': 'Rate per 100k',
            'Estimated_Cases': 'Est. Cases'
        })
    table_df['Est. Cases'] = table_df['Est. Cases'].apply(lambda x: f'{x:,}')
    table_df['Rate per 100k'] = table_df['Rate per 100k'].apply(lambda x: f'{x:,.1f}')

    table = dash_table.DataTable(
        data=table_df.to_dict('records'),
        columns=[{'name': c, 'id': c} for c in table_df.columns],
        sort_action='native',
        filter_action='native',
        page_size=10,
        style_table={'overflowX': 'auto'},
        style_header={'backgroundColor': NAVY, 'color': 'white', 'fontWeight': 'bold', 'fontSize': '13px'},
        style_cell={'fontSize': '13px', 'padding': '8px 12px', 'fontFamily': 'Arial'},
        style_data_conditional=[{'if': {'row_index': 'odd'}, 'backgroundColor': '#F8FAFC'}]
    )

    return cards, map_fig, bar_fig, trend_fig, region_fig, table


```

```python
# ── 4. RUN ─────────────────────────────────────────────────────────────────────

if __name__ == '__main__':
    app.run(debug=False)
```
